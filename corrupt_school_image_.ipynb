1. **Computer Vision in Fintech (Payments)**

Biometrics: Using facial recognition or fingerprints to verify people. eg.: unlocking my banking app with my face and confirming payment.

QR-Code Payments: Scanning codes instead of cash/cards. Computer vision reads the QR code and processes payment.

Swipe and Pay: Detecting my card or phone for payment, sometimes using cameras for secure authentication.


2. **Computer Vision in Healthtech**

Medical Imaging: Detecting diseases from X-rays, MRIs, or skin pictures automatically.

Remote Diagnosis: Mobile apps can take photos of symptoms (like rashes) and computer vision models can suggest what is wrong.

Monitoring Patients: Cameras in clinics or apps can track patient activity or adherence to treatment.
